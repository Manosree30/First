```python
missing_genres = df[df['song_name'].isnull()]['genre'].value_counts()
missing_genres.plot(kind='bar', figsize=(10,6), color='skyblue')
plt.title("Genres with Missing Song Names")
plt.xlabel("Genre")
plt.ylabel("Number of Missing Songs")
plt.xticks(rotation=45)
plt.show()
```

```python
df['genre'].duplicated().value_counts()
```

```python
df['genre'].value_counts().plot(kind='bar', figsize=(10,6), color='skyblue')
plt.xlabel("Genre")
plt.ylabel("Number of Songs")
plt.title("Number of Songs per Genre")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['danceability'], bins=20, color='skyblue', edgecolor='black')  # edgecolor adds borders
plt.xlabel('Danceability')
plt.ylabel('Frequency')
plt.title('Distribution of Danceability')
plt.show()
```

```python
df.groupby('genre')['danceability'].mean()
```

```python
#To see patterns in song lengths across genres.
df.groupby('genre')['duration_ms'].mean()
```

```python
df['duration_min'] = df['duration_ms'] / 60000
df.drop('duration_ms', axis=1, inplace=True)
df['duration_min']
```

```python
corr = df[['danceability','energy','valence']].corr()
print(corr)
```

```python
plt.figure(figsize=(6,5))
sns.heatmap(corr,annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation between Danceability, Energy, and Valence")
plt.show()
```

```python
df['tempo']
```

```python

df['tempo_category'] = pd.cut(df['tempo'], bins=[0, 80, 120, 2000], labels=['Slow','Medium','Fast'])
df['tempo_category']
```

```python
df['energy_dance'] = df['energy'] * df['danceability']
df['energy_dance']
```

```python
df['dance_happy'] = df['danceability'] * df['valence']
df['dance_happy']
```

```python
avg_energy = df.groupby('genre')['energy'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=avg_energy.index, y=avg_energy.values, palette='Reds_r', edgecolor='black')
plt.title("Average Energy by Genre (Top 10)")
plt.xlabel("Genre")
plt.ylabel("Average Energy")
plt.xticks(rotation=45)
plt.show()
```

```python
df.boxplot(column='danceability', figsize=(6,4))
plt.title("Boxplot of Danceability")
plt.xlabel("Danceability Feature")
plt.ylabel("Values")
plt.show()
```

```python
df.boxplot(column='tempo', by='genre', figsize=(10,6), rot=90)
plt.title("Boxplot of Tempo by Genre")
plt.suptitle("")
plt.xticks(rotation=45)
plt.show()
```

```python
df.groupby('genre')['danceability'].mean().sort_values(ascending=False).head(10).plot(kind='bar', figsize=(8,5))
plt.title("Average Danceability by Genre (Top 10)")
plt.ylabel("Danceability")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(df['loudness'], bins=30, kde=True, color='skyblue')
plt.title("Distribution of Loudness")
plt.xlabel("Loudness (dB)")
plt.ylabel("Number of Songs")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(df['speechiness'], bins=30, kde=True, color='lightcoral')
plt.title("Distribution of Speechiness")
plt.xlabel("Speechiness")
plt.ylabel("Number of Songs")
plt.show()
```
