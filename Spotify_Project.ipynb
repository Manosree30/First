```python
from sklearn.preprocessing import PowerTransformer

columns = ['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness',
           'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo',
           'time_signature', 'duration_min', 'energy_dance', 'dance_happy', 'is_acoustic']

# Apply Yeo-Johnson transformation
pt = PowerTransformer(method='yeo-johnson')
df[columns] = pt.fit_transform(df[columns])

# Plot to check results
import matplotlib.pyplot as plt
import seaborn as sns

n_cols = 4
n_rows = (len(columns) + n_cols - 1) // n_cols
fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, n_rows * 4))
axes = axes.flatten()
for i, col in enumerate(columns):
    sns.histplot(df[col], bins=30, kde=True, ax=axes[i])
    axes[i].set_title(f'Normalized {col}')
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])
plt.tight_layout()
plt.show()
```

```python
df['duration_min'] = df['duration_ms'] / 60000
df.drop('duration_ms', axis=1, inplace=True)
df['duration_min']
```

```python
df['tempo']
```

```python

df['tempo_category'] = pd.cut(df['tempo'], bins=[0, 80, 120, 2000], labels=['Slow','Medium','Fast'])
df['tempo_category']
```

```python
df['energy_dance'] = df['energy'] * df['danceability']
df['energy_dance']
```

```python
df['dance_happy'] = df['danceability'] * df['valence']
df['dance_happy']
```

```python
df['is_acoustic'] = df['acousticness'].apply(lambda x: 1 if x > 0.5 else 0)
df['is_acoustic']
```

```python

# Columns to plot (excluding 'genre' as it's categorical)
columns = ['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness',
           'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo',
           'time_signature', 'duration_min', 'energy_dance', 
           'dance_happy']

# Set figure size and number of rows and columns for subplots
n_cols = 4
n_rows = (len(columns) + n_cols - 1) // n_cols  # Ceiling division

fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, n_rows * 4))
axes = axes.flatten()

for i, col in enumerate(columns):
    sns.boxplot(data=df, y=col, ax=axes[i])
    axes[i].set_title(f'Box plot of {col}')

# Turn off empty subplots
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])
    
plt.tight_layout()
plt.show()
```

```python
numeric_cols = ['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness',
                'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo',
                'time_signature', 'duration_min', 'energy_dance', 'dance_happy']

Q1 = df[numeric_cols].quantile(0.25)
Q3 = df[numeric_cols].quantile(0.75)
IQR = Q3 - Q1

for col in numeric_cols:
    lower_bound = Q1[col] - 1.5 * IQR[col]
    upper_bound = Q3[col] + 1.5 * IQR[col]

    df[col] = np.where(df[col] < lower_bound, lower_bound, df[col])
    df[col] = np.where(df[col] > upper_bound, upper_bound, df[col])

n_cols = 4
n_rows = (len(numeric_cols) + n_cols - 1) // n_cols

fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, n_rows * 4))
axes = axes.flatten()

for i, col in enumerate(numeric_cols):
    sns.boxplot(data=df, y=col, ax=axes[i])
    axes[i].set_title(f'Box plot of {col} (Capped)')

for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```
